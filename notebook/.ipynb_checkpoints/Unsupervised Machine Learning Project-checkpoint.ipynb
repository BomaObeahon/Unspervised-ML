## Principal Component Analysis

```python
features= ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
X = data[features]
```

```python
X
```

```python
# Standardize numerical variables

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Scaling a dataset transforms the features (columns) of the dataset so that they have a similar scale or range of values. This transformation is important because many machine learning algorithms perform better when the features are on a similar scale.

```python
scaled_df = pd.DataFrame(X_scaled)
scaled_df
```

```python
# Performing PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X_pca = pca.fit_transform(scaled_df)
```

```python
# Here it reduced the 6 features of the dataset to 2 features
print(scaled_df.shape)
print(X_pca.shape)
```

```python
pca.explained_variance_ratio_
```

```python
# Output explanation
print("Principal Components (Explained Variance Ratio):")
for i, explained_variance_ratio in enumerate(pca.explained_variance_ratio_):
    print(f"Principal Component {i+1}: {explained_variance_ratio:.2f}")
```

```python
# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.scatter(x=[i+1 for i in range(len(pca.explained_variance_ratio_))],
            y=pca.explained_variance_ratio_,
            s=200, alpha=0.75, c='orange', edgecolor='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n", fontsize=15)
plt.xlabel('Principal Component', fontsize=10)
plt.ylabel('Explained Variance Ratio', fontsize=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Plotting scatterplot with principal components
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1],c=data['Channel'],cmap='plasma', edgecolors='k',alpha=0.75,s=150)
plt.xlabel("Principal component-1", fontsize=10)
plt.ylabel("Principal component-2", fontsize=10)
plt.title("Class separation using first two principal components\n", fontsize=15)
plt.show()
```

Clearly by using these 2 components we can easily seperate the two classes.

### Interpreting the components

Unfortuately, being able to easily understand what these component represent is not possible.

These components corresponds to combination of the original features, the componenets themselves are stored as an attribute of the fitted PCA object.

```python
#pca_components
```

In this numpy matrix array,each row represents a principal component, and each relates back to the original features. We can visualize this relationship using a heatmap.

```python
data_comp = pd.DataFrame(pca.components_, columns= X.columns)
plt.figure(figsize=(12,6))
sns.heatmap(data_comp,cmap='plasma');
```

This heatmap and the color bar represent the correlation between the different feature and the principal component itself.

The output "Principal Components (Explained Variance Ratio)" indicates the proportion of variance in the original dataset explained by each principal component obtained through PCA.

Principal Component 1 explains approximately 44% of the variance in the dataset.
Principal Component 2 explains approximately 28% of the variance in the dataset.

The explained variance ratio tells how much information (variance) is retained in each principal component compared to the original dataset. In this case, Principal Component 1 captures the highest amount of variance, followed by Principal Component 2.

 A higher explained variance ratio indicates that the principal component retains more information from the original dataset. Therefore, in this example, Principal Component 1 captures more variability in the data compared to Principal Component 2.

## KMeans Clustering 

```python
# Determine the Optimal Number of Clusters

wcss = []  # within cluster sum of squares(wcss)
for i in range(1,12):
    kmeans = KMeans(n_clusters = i, n_init= 10, random_state = 42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,12), wcss, color = 'red', marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show();
```

It is best practice to allow the algorithm choose the optimal number of clusters required when using K-Means in Unsupervised Machine Learning, this can be done by using the Elbow Method. The optimal number of cluster is important because it affects the quality of the clustering results.
WCSS(within cluster sum of squares) is used to determine the optimal number of clusters.

The approach consists of looking for a kink or elbow in the WCSS graph. Usually, the part of the graph before the elbow would be steeply declining, while the part after it â€“ much smoother. In this instance, the kink comes at the 5 clusters mark. So, weâ€™ll be keeping a five-cluster solution.



```python
# Plot the generated data
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=5)
plt.show()
```

```python
# Initialize K-means with random centroids
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)  # Set n_init to the desired number of initializations
```

```python
# Perform K-means clustering
kmeans.fit(X_scaled)
```

```python
# Perform K-means clustering
y_pred = kmeans.predict(X_scaled)
y_pred
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_pred, s=5)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, y_pred)
```

silhouette score of 0.3549546531235654 suggests that the clustering algorithm has produced moderately well-separated clusters, with some overlap between clusters but overall decent cohesion within each cluster.
